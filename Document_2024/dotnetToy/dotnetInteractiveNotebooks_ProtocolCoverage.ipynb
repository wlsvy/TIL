```polyglot-notebook
using static System.Console;
using System;
using System.IO;

//string filePath = "d:\\MM/Sources/TestClient/TestEnvironment/Networking/CG/AutoGenerated/CGRpcRequester.g.cs";
string filePath = "d:\\MM/Sources/Common/Shared/Protocols/CG/AutoGenerated/CGNotificationSender.g.cs";
string fileText = File.ReadAllText(filePath);
//string startWith = "public async Task<";
//string prefixToParse = "Request_";
string startWith = "public void Notify";
string prefixToParse = "Notify_";

string Preprocess(string singleLine)
{
    int index = singleLine.IndexOf(startWith);
    if(index == -1){
        return string.Empty;
    }

    int start = singleLine.IndexOf(prefixToParse);
    var end = singleLine.IndexOf('(');
    if(start == -1 || end == -1)
    {
        return string.Empty;
    }
    return singleLine[start..end].Trim();
}


string[] ToStrings(string lines)
{
    var ret = new List<string>();
    foreach(var line in lines.AsSpan().EnumerateLines())
    {
        if(line.IsEmpty)
        {
            continue;
        }
        ret.Add(line.ToString());
    }
    
    return ret.ToArray();
}



var newCol = ToStrings(fileText)
    .Select(x => Preprocess(x))
    .Where(x => !string.IsNullOrEmpty(x))
    .ToArray();
var oldCol = ToStrings(oldList);
WriteLine("===========NewValues===========");
WriteLine(string.Join("\n", newCol.Take(5)));
WriteLine($"...");
WriteLine($"Total: [{newCol.Length}]");
WriteLine("===========OldValues===========");
WriteLine(string.Join("\n", oldCol.Take(5)));
WriteLine($"...");
WriteLine($"Total: [{oldCol.Length}]");

WriteLine("===========Newly Added===========");
WriteLine(string.Join("\n", newCol.Except(oldCol).Distinct().OrderBy(x => x)));
WriteLine("===========Removed===========");
WriteLine(string.Join("\n", oldCol.Except(newCol).Distinct().OrderBy(x => x)));
WriteLine("======================");
```
